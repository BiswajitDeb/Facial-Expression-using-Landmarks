```python
import pandas as pd
df = pd.read_csv("Affectnet Dataset/labels.csv")
df.sample(5)
```

```python
df.info
```

```python
import cv2
import mediapipe as mp
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh()
```

```python
X_preprocessed = []
Y_preprocessed=[]
lost=0
```

```python
for i in range(0,28175):
    img = cv2.imread("Affectnet Dataset/"+df.iloc[i]['pth'])
    result = face_mesh.process(img)

    if result.multi_face_landmarks is not None:
        landmarks_per_image = []
        for facial_landmarks in result.multi_face_landmarks:
            for landmark in facial_landmarks.landmark:
                landmarks_per_image.append((landmark.x, landmark.y))
    else:
        lost+=1
    print(f"Image : {i} Lost : {lost}")
    
    X_preprocessed.append(landmarks_per_image)
    Y_preprocessed.append(df.iloc[i]['label'])
```

```python
import pandas as pd
df_landmarks = pd.DataFrame(X_preprocessed)
df_landmarks.to_csv('Saved_landmarks1.csv', index=False)
```
